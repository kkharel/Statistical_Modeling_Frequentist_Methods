# **HYPOTHESIS TESTING**



**Hypothesis testing** is a method in statistics that helps us make decisions or draw conclusions using data.

We start with a claim that is assumed to be true unless there is a strong evidence against it. This is called a Null Hypothesis (**$H_0$**). Then we consider an alternative claim we want to test.This is called the Alternative Hypothesis (**$H_1$**). Then we collect the data and check whether the evidence is strong enough to reject the null hypothesis.

The goal in hypothesis testing is to **Try to find evidence against null $H_0$**.


## **Types of Hypotheses**:

1. **Null Hypothesis $H_0$**: The default assumption or statement of **no effect / no difference**. It always includes equality. For example: The new drug has **no effect** on blood pressure

2. **Alternative Hypothesis $H_1$**: The statement we want to find evidence for. It represents **an effect, difference, or relationship**.  

**Types of alternative hypotheses:**

1. **Two-tailed (non-directional)**: Tests if the parameter is **different** (could be higher or lower).  For example: $H_0: \mu = 100$, $H_1: \mu \neq 100$. Two tailed test is used when we only care about if there is a difference, not the direction.

2. **One-tailed (directional)**: Tests if the parameter is **greater or smaller** than a value. For example (upper-tail): $H_0: \mu = 100$, $H_1: \mu > 100$, (lower-tail): $H_0: \mu = 100$, $H_1: \mu < 100$. One tailed test is used when we have a specific expectation about the direction of the effect.


**H0 always includes equality** (=). **H1 always includes inequality** $(\neq, >, <)$ . The choice of one-tailed vs two-tailed affects **critical value**, **p-value**, and **power**. In practice: if we only care about an increase or decrease, use **one-tailed** and if we care about any difference, use **two-tailed**.



Let's look at the example: In a court of law **A person is assumed innocent until proven guilty**.

**Null Hypothesis $H_0$:** A person is innocent

**Alternative Hypothesis $H_1$:** A person is guilty (this is what the prosecutor is trying to prove)

| Courtroom                               | Statistics                                      |
|-----------------------------------------|--------------------------------------------------|
| Assume innocence                        | Assume $H_0$ is true                           |
| Present evidence against the defendant  | Collect data to test $H_0$                     |
| If evidence is strong → convict         | If data strongly contradicts $H_0$ → reject $H_0$ |
| If evidence is weak → acquit            | If data does not contradict $H_0$ enough → fail to reject $H_0$ |




The outcomes of the case can be as follows:

|                                | **Predicted: Fail to Reject $H_0$** | **Predicted: Reject $H_0$** |
|-------------------------------|---------------------------------------|-------------------------------|
| **Actual: $H_0$ True**      | **TN** (True Negative)                | **FP** (False Positive) = Type I Error |
| **Actual: $H_1$ True**      | **FN** (False Negative) = Type II Error | **TP** (True Positive)        |


- courtroom predicts guilty and the person is acutally guilty
- courtroom predicts not guilty and the person is actually not guilty
- courtroom predicts guilty and the person is not actually guilty
- courtroom predicts not guilty and the person is actually guilty.

When the courtroom predicts guilty when the person is actually not guilty then this is a **FP** that is **Type I** error meaning that a innocent person is convicted. The probability of **Type I** error is denoted by alpha ($α$) and is typically set as the significance level (e.g., 0.05) in hypothesis testing. Minimizing **Type I** error is important when the cost or consequences of making a false positive decisions are high.

When the courtroom predicts not guilty when the person is actually guilty then this is a **FN** that is **Type II** error meaning that a guilty person is not convicted. The probability of **Type II** error is denoted by beta ($β$). Minimizing **Type II** error is crucial when failing to detect a real effect or difference can have significant implications. In our case, β is the chance we fail to detect a real effect and letting a guilty person go free.

**How $β$ is chosen?**

Unlike α (Type I error), which is usually fixed (0.05 or 5%), **β** is related to the **power** of a test. Recall that $\beta$ is the probability of failing to reject $H_0$ when $H_1$ is true. The power of the test is given below,

$$
\text{Power} = 1 - \beta
$$

which is the probability of correctly rejecting $H_0$ when $H_1$ is true.

Minimizing **β** (or increasing power) is important when missing a real effect has serious consequences, e.g.,:

- **Medical tests:** failing to detect a disease  
- **Quality control:** missing a defective product  

The choice of **β** depends on context:

- If consequences of Type II error are severe then choose smaller β (higher power).  
- Typical β values are 0.1, 0.2 (corresponding to power of 90% or 80%).

**What is p-value?**

**p-value** which represents the probability of obtaining the results as extreme as those observed, assuming the null hypothesis is true. Smaller p-values indicate stronger evidence against the null hypothesis.

Suppose we have a magic hat with red balls and blue balls. We believe that there are 50% red balls and 50% blue balls in the hat (this is our null hypothesis). Now, we pull out 10 balls from the hat and 9 of them are blue balls. The p-value is like asking **Hmm... If the hat really had 50/50 balls, how surprising is it to pull out 9 blue balls?**. If it is very surprising then maybe the hat is not fair and we don't believe our initial guess anymore. If it is not surprising then the hat might be fair and our initial guess is ok. If **p-value** is small then it is very surprising hence our null maybe wrong. If it is big then it is not very surprising hence our null maybe true.

## **Steps for Hypothesis Testing (from EDA to Decision)**

1) **Exploratory Data Analysis (EDA)**
- **Goal:** Understand the data before forming hypotheses.
- Check data types, distributions, patterns, and possible outliers.
- This step is used to **generate ideas and potential hypotheses**, not to formally test them.

2) **Formulate Hypotheses**
- Define **Null Hypothesis $H_0$**: the default assumption (e.g., “the coin is fair” or “the drug has no effect” or "the person is innocent").
- Define **Alternative Hypothesis $H_1$**: what we are trying to find evidence for (e.g., “the coin is biased” or “the drug works” or "the person is guilty").

3) **Set Significance Level (α) and Type II Error (β)**
- **α (Type I error):** Probability of rejecting $H_0$ when it is actually true. Commonly 0.05 (5%).  
- **β (Type II error):** Probability of failing to reject $H_0$ when $H_1$ is true.  
- **Power = 1 - β:** Probability of correctly rejecting $H_0$ when $H_1$ is true.  
- Choose α and β based on how severe each error would be in our context.

4) **Plan and Collect Data**
- Collect new data specifically for **testing hypotheses**.
- Ensure sample size is sufficient to achieve the desired **power**.

5) **Choose Appropriate Statistical Test**
- Decide the test based on **data type** and **research question**:
  - **t-test** → Compare means of two groups
  - **ANOVA** → Compare means of multiple groups
  - **Chi-square** → Test for association between categorical variables
  - Others depending on context

6) **Perform the Test and Calculate P-value**
  - **P-value:** Measures how surprising the data is if $H_0$ is true.
    - **Small p-value** → very surprising → evidence against $H_0$  
    - **Large p-value** → not surprising → not enough evidence to reject $H_0$

7) **Make a Decision**
- Compare **p-value** to **α**:
  - **p-value < α** → Reject $H_0$ (evidence supports $H_1$)  
  - **p-value ≥ α** → Fail to reject $H_0$ (not enough evidence)
- Consider **β / power** if relevant for risk of Type II error.

8) **Interpret Results in Context**
- Translate statistical decision into **real-world implications**:
  - What does rejecting or failing to reject $H_0$ mean for research question?
  - Consider consequences of Type I and Type II errors.

## **How to determine Power and Sample Size?**

**Power (1-β)** is the probability of correctly rejecting $H_0$ when $H_1$ is true.  

Power depends on three factors:

1. **Effect Size (d)**: the difference we want to detect relative to the variability.  
   $$
   d = \frac{\text{Difference in means}}{\text{Standard deviation}}
   $$

2. **Sample Size (n)**: number of observations.  
   - As n gets larger, standard error gets smaller which implies that it is easier to detect the effect meaning we have higher power.

3. **Significance Level (α)**: probability of Type I error (commonly 0.05).  


**Step 1: Specify Parameters:**

Suppose we want to test if a new teaching method increases test scores.

- Null hypothesis $H_0$: Mean score = 70  
- Alternative hypothesis $H_1$: Mean score > 70  
- Standard deviation $\sigma$ = 10  
- Expected improvement (effect size) = 5 points  
- Significance level $\alpha$ = 0.05  
- Desired power = 80% → β = 0.2


**Step 2: Calculate Effect Size:**

$$
d = \frac{\text{Difference in means}}{\text{Standard deviation}} = \frac{5}{10} = 0.5
$$

**Step 3: Determine Required Sample Size:**

For a **one-sample z-test**:

$$
n = \left( \frac{(Z_\alpha + Z_\beta) \cdot \sigma}{\text{Effect Size}} \right)^2
$$

where:  
- $Z_\alpha$ is the critical value for significance level α (e.g., 1.645 for one-tailed 0.05)  
- $Z_\beta$ is the critical value corresponding to power (e.g., 0.84 for 80% power)  
- σ = standard deviation  
- Effect Size = difference to detect

Plug in the numbers:

$$
n = \left( \frac{(1.645 + 0.84) \cdot 10}{5} \right)^2
 \approx 24.7
$$

Hence, need **at least 25 students** to detect a 5-point improvement with 80% power at 5% significance.

**Step 4: Interpret**

- If we collect **25 or more students**, study has enough power to detect a 5-point improvement.  
- If we collect fewer students, power decreases and there is a higher chance of **Type II error (β)**.

**Notes**:

When we compare two groups (like a treatment group and control group), many basic statistical tests assume that the two groups have the same amount of variability. In other words, they assume the **variances are equal**. This is the reason why the formulas for two-sample t-tests often use a **pooled standard deviation**. A pooled standard deviation is just a single number that combines the spread of both groups under the assumption that their spreads are similar.



```python
from statsmodels.stats.power import TTestPower

# Parameters
effect_size = 5 / 10  # difference / std
alpha = 0.05
power = 0.8

# Calculate required sample size
analysis = TTestPower()
n_required = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')
print(f"Required sample size: {n_required:.0f}")
```

Manual calculation uses a z-test (assumes the true standard deviation is known), while TTestPower uses a t-test (assumes the standard deviation must be estimated). The t-test is slightly more conservative for small samples, so it gives a slightly larger sample size (26 instead of 25).

## **Types of Statistical Tests with Applications**

Statistical tests help us decide whether to reject $H_0$ based on our data. The choice of test depends on:

- **Type of data** (categorical, numerical)  
- **Number of groups or variables**  
- **Distribution of data** (normal or non-normal)  

**t-Test**
- **Purpose:** Compare the means of two groups.  
- **Types:**
  1. **One-sample t-test**: Compare sample mean to a known value.  
     - **Example:** Is the average test score different from 70?  
     - **Criteria:** Numerical data, one group, normally distributed  
     - **Applications:** Employee satisfaction surveys, average sales per store
  2. **Independent two-sample t-test**: Compare means of two independent groups.  
     - **Example:** Compare scores of boys vs girls.  
     - **Criteria:** Numerical data, two independent groups, normally distributed  
     - **Applications:** Compare performance of two marketing campaigns, test scores across departments
  3. **Paired t-test**: Compare means of related groups (before-after).  
     - **Example:** Measure weight before and after a diet.  
     - **Criteria:** Numerical data, paired/repeated measures, normally distributed  
     - **Applications:** Effect of training programs, A/B testing results

**ANOVA (Analysis of Variance)**
- **Purpose:** Compare means of three or more groups.  
- **Example:** Compare test scores among students from three different schools.  
- **Types:**  
  - **One-way ANOVA:** One factor (e.g., school)  
  - **Two-way ANOVA:** Two factors (e.g., school and gender)  
  - **MANOVA (Multivariate ANOVA):** Tests multiple dependent variables.

- **Criteria:** Numerical data, 3+ groups, independent samples, normal distribution, equal variances  
- **Applications:** Product quality comparison across factories, marketing spend impact on regions, clinical trial outcomes

**Chi-Square Test**
- **Purpose:** Test association or independence between categorical variables.  
- **Example:** Is smoking habit related to gender?  
- **Types:**  
  - **Chi-Square Goodness-of-Fit Test:** We use this when we have one categorical variable and we want to test if the distribution matches expected proportions.
  -  **Chi-Square Test of Independence:** We use this when we have two categorical variables and want to test if they are independent
  - **Chi-Square Test for Homogeneity:** We use this when we compare categorical distributions across groups


- **Criteria:** Categorical data, expected frequency per cell ≥ 5, independent observations  
- **Applications:** Customer preferences by region, survey responses, brand choice behavior.

**Non-Parametric Tests**
- **Purpose:** For data that doesn't follow a normal distribution.  
- **Examples:**
  - **Mann-Whitney U test:** Compare two independent groups  
  - **Wilcoxon signed-rank test:** Compare two related groups  
  - **Kruskal-Wallis test:** Compare more than two groups  
- **Criteria:** Numerical or ordinal data, small sample size, not normally distributed, independent or paired as required  
- **Applications:** Customer ratings, survey Likert scales, clinical trial ordinal outcomes

**Other Tests**
- **F-test:** Compare variances of two groups  
  - **Applications:** Variability in production lines, financial volatility comparison  
- **Z-test:** Like t-test, used for large samples or known population variance  
  - **Applications:** Proportion testing in surveys, quality control metrics  
- **Logistic regression:** Predict binary outcome (yes/no)  
  - **Applications:** Churn prediction, loan default, click-through rate analysis

**Notes:**

For tests like t-test and ANOVA, the true requirement is that the sample means are normally distributed. This comes from the Central Limit Theorem which states : When the sample size is large enough, the distribution of sample mean becomes approximately normal even if the raw data is not normally distributed. The raw data must be normal only when the sample size is small (n < 30).

## **Observational Study vs Experimental Study**

Hypothesis testing works the same in both study types. But the study design affects interpretation not the mechanics. In experiments, a significant result imples **causation** while in observational studies, a significant result imples only **association**. If the study is observational, we cannot claim causation even if the test is significant.

Observational study is a type of research where the investigator simply observes and records data without interfering or assigning treatments. The researcher does not control who receives what condition, they just study variables as they naturally occur. Because of this lack of control, observational studies can identify associations between variables but cannot establish causation, since confounding factors may influence the results.

In contrast, an experimental study involves deliberately manipulating one or more variables and assigning subjects to different groups, often using randomization. This controlled setup allows researchers to isolate the effect of the variable of interest, making it possible to draw stronger conclusions about **cause and effect**.

To summarize, observational studies reveal patterns while experimental studies can test causal relationships.

## **Why Observational vs Experimental study Matters?**

There are several observational study threats:

1) **Confounding**: A third variable affects both variables.

2) **Reverse causality**: Outcome might affect the “cause.”

3) **Selection bias**: Sample not random.

4) **Omitted variable bias**: Missing predictors.

5) **Measurement error**: Inaccurate data.

6) **Ecological fallacy**: Area-level data may mislead about individuals.

Experimental studies avoid most of the above. Randomization balances confounders (both observed and unobserved).

I will discuss about how to make observational analysis stronger while talking about applications below. I will also talk about quasi-experimental when randomization is not possible.

# **APPLICATIONS**

## **Observational Pre-Post Study**

Suppose we collected revenue data from 30 sellers on our platform before and after they advertised their products using a cost-per click model. We want to know

**Research Question: "Is there an increase in revenue after sellers advertise?"**

This is observational study not experimental since we did not assign who advertises. we did not control the intervention, we cannot say the revenue lift was caused by the advertising model. Other factors could have affected revenue during the same period (seasonality, promotions, market demand etc.). Even though this is before-after for the same sellers, this is still pre-post observational study, not a randomized experiments. We measure association here not causation. The design is pre-post paired study.

**Hypotheses**

Let $\mu_{after}$ = average revenue after using the cost-per-click advertising

and $\mu_{before}$ = average revenue before using the advertising

We are interested in the mean difference in revenue for sellers:

$$\mu_{d} = \mu_{after} - \mu_{before}$$

The null hypothesisi is as follows:

$H_0: \mu_d = 0$, No Change in Revenue

The alternative hypothesis is as follows:

$H_0: \mu_d > 0$, Revenue increased after advertising

**Test Type: paired t-test, one-tailed**

**Assumptions:** For a paired t-test, the differences are approximately normally distribtuted. Observations (sellers) are independent.

**Data Simulation and Visualization**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n_sellers = 30
pre_revenue = np.random.normal(loc=70, scale=8, size=n_sellers)
post_revenue = pre_revenue + np.random.normal(loc=5, scale=5, size=n_sellers)
pre_post_df = pd.DataFrame({'Pre': pre_revenue, 'Post': post_revenue})

# Visualizations
sns.histplot(pre_post_df['Pre'], color='blue', kde=True, label='Pre', alpha=0.5)
sns.histplot(pre_post_df['Post'], color='red', kde=True, label='Post', alpha=0.5)
plt.title("Pre vs Post Revenue Distribution")
plt.legend()
plt.show()

sns.boxplot(data=pre_post_df)
plt.title("Pre vs Post Revenue Boxplot")
plt.show()
```

Above histogram and kde plot for pre and post revenue shows overall distributions. boxplot shows median and spread, highlighting the increase after advertising.

**Check Normality of Differences**

```python
from scipy.stats import shapiro

differences = pre_post_df['Post'] - pre_post_df['Pre']
stat, p = shapiro(differences)
print(f'Shapiro-Wilk test: stat={stat:.3f}, p={p:.3f}')

# p > 0.05 implies differences roughly normal, it is safe to use paired t-test. If p was <= 0.05 then
# differences would have deviated from normal and we had to consider non-parametric approaches (Wilcoxon signed-rank test)
```

t-test is valid. Note that we do not need pre and post to have equal variances, only the differences matter.

### **paired t-test**

```python
from scipy import stats

t_stat, p_val = stats.ttest_rel(pre_post_df['Post'], pre_post_df['Pre'])
# # we are testing post > pre hence one-tailed test

p_one_tailed = p_val / 2 if t_stat > 0 else 1 - (p_val / 2)

print(f'Paired t-test: t = {t_stat:.3f}, p (one-tailed) = {p_one_tailed:.3f}')
```

we performed a paired t-test to see if the mean difference in revenue is significantly greater than zero. Since p_value is less than 0.05, we reject the null and conclude that revenue after advertising is significantly higher than before amonth the 30 sellers observed.

**effect size (cohen's d)**

```python
cohens_d = differences.mean() / differences.std(ddof=1)
print(f"Cohen's d: {cohens_d:.3f}")
```

Cohen's d measures the magnitude of the effect, independent of sample size. It tells us how meaningful or large the effect is in real life.

$$
d = \frac{MeanDifference}{SDofDifferences}
$$

If it is around 0.2 then small effect, 0.5 then medium effect and 0.8 and above is large effect.

**Note: Statistical significance tells us whether an effect is likely due to random chance or not. Practical significance tells us how meaningful or large the effect is in real life. Even a tiny effect can be statistically significant if sample is huge but it might not matter practically.**

From our results it indicates a large effect, meaning the revenue lift is substantial in practice, not just statistically significant. Advertising really makes a noticeable increase in seller revenue, not just a tiny statistical blip.

**post-hoc power (optional for observational study)**

```python
from statsmodels.stats.power import TTestPower

analysis = TTestPower()
power = analysis.solve_power(effect_size=cohens_d, nobs=n_sellers, alpha=0.05, alternative='larger')
print(f"Achieved power: {power:.3f}")
```

Power is the probability of detecting an effect if it truly exists, it is especially useful for small sample studies to understand the reliability of the test. With the sample size of 30 and effect size of 0.94, our test is extremely likely to detect the observed revenue increase.

**Note:** In observational studies, power analysis is descriptive, it does not make results causal.

**Conclusion**:

Revenue after advertising is significantly higher than before. The magnitude of increase is large. Our study has high statistical power meaning the observed difference is very unlikely to be due to random chance. **Because this is an observational study, we can only say that advertising is associated with revenue increase, not that it caused it. Other factors like market trends, seasonality etc.. could also contribute.**

### **one sample t-test**

Suppose we want to check if the average pre-advertising revenue of sellers is different from a known target revenue of 70 then we would conduct observational study.

$H_0: mean revenue = 70$

$H_1: mean revenue \neq 70$

```python
from scipy.stats import ttest_1samp

t_stat, p_val = ttest_1samp(pre_revenue, popmean=70)
print(f"One-sample t-test: t = {t_stat:.3f}, p = {p_val:.3f}")
```

Here, t_stat measures how far the sample mean is from the population mean in units of standard error. p value tells us the probabilty of observing this t-statistic if the true mean were 70. In our case, we fail to reject the null hypothesis and conclude the average pre-advertising revenue of sellers is not significantly different from the target value of 70.

```python
cohens_d = (pre_revenue.mean() - 70) / pre_revenue.std(ddof=1)
print(f"Cohen's d = {cohens_d:.3f}")
```

Even though there is a small negative difference (mean slightly below 70), it is not practically meaningful. Statistically, the difference might exist but it is too tiny to impact passing/failing decisions. p-value tells us if it's real and effect size tells us if it matters.

Now, let's turn our focus toward experimental study.


### **independent t-test**

```python
import numpy as np
from scipy import stats

np.random.seed(42)
n_treat = 15
n_control = 15

# Observational: naturally occurring treatment
treatment_revenue = np.random.normal(75, 8, n_treat)
control_revenue = np.random.normal(70, 8, n_control)

t_stat, p_val = stats.ttest_ind(treatment_revenue, control_revenue, alternative='greater')
print(f"Observational independent t-test: t={t_stat:.3f}, p-one-tailed={p_val:.3f}")
```

p < 0.05 implies revenue is higher in the treatment group, but we cannot claim causation, only association.

### **Difference-in-Differences (DiD)**

```python
# Pre-post revenues
pre_treatment = np.random.normal(70, 8, 15)
post_treatment = pre_treatment + np.random.normal(5, 5, 15)  # observed change

pre_control = np.random.normal(70, 8, 15)
post_control = pre_control + np.random.normal(2, 5, 15)      # general trend

# Change scores
change_treatment = post_treatment - pre_treatment
change_control = post_control - pre_control

t_stat, p_val = stats.ttest_ind(change_treatment, change_control, alternative='greater')
print(f"Observational DiD t-test: t={t_stat:.3f}, p-one-tailed={p_val:.3f}")
```

This test measures whether the treatment group increased more than control, controlling for general trends. It is still associational, not strictly causal.

We can do t-test like above but more commonly we use regression-based DiD to adjust for confounders. T-tests on simple change score are possible if groups looks balanced and parallel trend assumption holds approximately.

## **Experimental Between-Subjects Study (Randomized Controlled Design)**

**independent t-test**

**Research question: Does using cost-per-click advertising model increase revenue for sellers on our platform?**

What are the key differences from here from observational study?

We now assign sellers randomly to advertise (treatment) or not (control). This allows for causal inference, unlike the observational case.


**Study Design**:
- Subjects: 30 sellers
- Groups: Randomly assign 15 sellers to advertising (treatment), 15 to no advertising (control)
- Outcome: Revenue after 30 days
- Analysis: Compare the average revenue between the two groups.

Randomization ensures that other factors (e.g., seasonality, seller skill) are evenly distributed, isolating the effect of advertising.

**Hypothesis Formulation**:

Let $\mu_{treatment}$ = mean revenue of sellers who advertise

and let $\mu_{control}$ = mean revenue of sellers who do not advertise

$H_0: \mu_{treatment} - \mu_{control} = 0$, no difference in revenue

$H_1: \mu_{treatment} - \mu_{control} > 0$, advertising increases revenue

**test type: independent two sample t-test (one-tailed)**

**Assumptions**:
- Independence: Revenue for each seller is independent
- Normality: Revenue in each group is roughly normal
- Equal Variances (optional): Can use Welch's t-test if variances differ

**Simulate example data**

```python
import numpy as np
import pandas as pd

np.random.seed(42)
n_sellers = 30

# Randomly assign 15 to treatment, 15 to control
treatment_revenue = np.random.normal(loc=75, scale=8, size=15)  # after advertising
control_revenue = np.random.normal(loc=70, scale=8, size=15)    # no advertising

exp_df = pd.DataFrame({'Revenue': np.concatenate([treatment_revenue, control_revenue]),'Group': ['Treatment']*15 + ['Control']*15})
exp_df.head()
```

**check normality**

```python
stat_t, p_t = shapiro(treatment_revenue)
stat_c, p_c = shapiro(control_revenue)
print(f"Treatment p={p_t:.3f}, Control p={p_c:.3f}")

# p > 0.05, roughly normal, independent t-test valid
```

Both treatment and control group revenues appear roughly normally distributed.This satisfies the normality assumption for an independent two-sample t-test

**check variances**

```python
treatment_var = np.var(treatment_revenue, ddof=1)
control_var = np.var(control_revenue, ddof=1)

print(f'Treatment variance: {treatment_var:.2f}')
print(f'Control variance: {control_var:.2f}')
```

**Rule of thumb**: If the ratio of the larger variance to the smaller variance is less than 4 then the standard t-test is generally robust. If we want to be extra cautious we can use Welch's t-test which does not assume equal variances.

The variance ratio (63.25 / 38.73 ≈ 1.63) is below the common robustness threshold of 4, indicating that the equal-variance assumption is reasonably met. Thus, a standard two-sample t-test is appropriate, though we also report Welch's t-test for completeness.

### **independent t-test**

```python
t_stat, p_val = stats.ttest_ind(treatment_revenue, control_revenue, alternative='greater')
print(f"Independent t-test: t={t_stat:.3f}, p-one-tailed={p_val:.3f}")
```

```python
t_stat_welch, p_val_welch = stats.ttest_ind(treatment_revenue, control_revenue, alternative='greater', equal_var=False)
print(f"Welch's t-test: t = {t_stat_welch:.3f}, p = {p_val_welch:.3f}")
```

**Decision:** Reject $H_0$ since p is less than 0.05 and conclude "advertising caused revenue increase".

**Because this is an experimental study, we can make a causal statement: "Advertising caused revenue to increase."**

**effect size (cohen's d)**

For independent groups:

$$
d = \frac{mean_{treatment} - mean_{control}}{pooledSD}
$$

```python
pooled_sd = np.sqrt(((14* treatment_revenue.std(ddof=1)**2) + (14* control_revenue.std(ddof=1)**2)) / 28)
cohens_d = (treatment_revenue.mean() - control_revenue.mean()) / pooled_sd
print(f"Cohen's d = {cohens_d:.3f}")
```

This helps interpret practical significance, independent of sample size. The effect is very large, indicating a meaniful revenue increase.

**power analysis (Pre-Study Planning)**

In experimental studies, power analysis is used before the study to determine sample size needed.

```python
from statsmodels.stats.power import TTestIndPower

effect_size = cohens_d
analysis = TTestIndPower()
required_n = analysis.solve_power(effect_size=effect_size, alpha=0.05, power=0.8, alternative='larger')
print(f"Required sample size per group: {required_n:.0f}")
```

This ensures the study is likely to detect a true effect. The required sample size per group for 80% power is 10. Achieving higher power means that we are very likely to detect true effects.

**80% power means that if the advertising truly increases revenue, study has an 80% chance of detecting that increase.**

**There is a 20% chance (β = 0.20) that the study will miss a real effect even if it exists.**

**Conclusion:**
Independent t-test shows a statistically significant increase in revenue for sellers who advertise. Cohen's d indicates the increase is large and meaningful in practice. Random assignment allows a causal conclusion: "Advertising caused revenue to increase among sellers in this study." Sample size and power confirm the study was adequately designed to detect the true observed effect.

## **Experimental Pre-Post Design with Control (Difference-in-Differences)**

**Research Question: Does cost per click advertising increase revenue for sellers compared to sellers who do not advertise?**

**Study Design**

- **Subjects:** 30 sellers

- **Groups:** Randomly assign 15 to Treatment (advertising), 15 to Control (no advertising)

- **Measurements:**
  - Pre: revenue for 30 days before advertising
  - Post: revenue for 30 days after advertising

- **Outcome:** revenue change for each seller = Post-Pre

- **Analysis:** compare average change between treatment and control: this is Difference-in-Differences (DiD)

**This design controls for general trends (e.g., seasonality, market demand) that affect both groups.**

**Hypothesis**

$H_0: \Delta_{treatment} - \Delta{control} = 0$ (no effect)

$H_1: \Delta_{treatment} - \Delta{control} > 0$ (advertising increases revenue)

**Pre-Study Power & Sample Size Planning**

Suppose we expect: Treatment effect = 4 units (additional revenue due to ads) and Standard deviation of change scores ≈ 5 units then

Effect size (Cohen's d):

$$
d = \frac{MeanDifferenceBetweenGroups}{pooledSD} = 4/5=0.8
$$

**Target: α = 0.05, power = 0.8, calculate required sample size per group:**

```python
from statsmodels.stats.power import TTestIndPower

effect_size = 0.8
alpha = 0.05
power_target = 0.8

analysis = TTestIndPower()
required_n = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power_target, alternative='larger')
print(f"Required sample size per group: {required_n:.0f}")
```

```python
# simulate dataset after sample size planning

import numpy as np
import pandas as pd

np.random.seed(42)
n_per_group = 25

# Pre-revenue
pre_treatment = np.random.normal(70, 8, n_per_group)
pre_control = np.random.normal(70, 8, n_per_group)

# Post-revenue
post_treatment = pre_treatment + np.random.normal(5, 5, n_per_group)  # treatment effect
post_control = pre_control + np.random.normal(1, 5, n_per_group)      # general trend

# Change scores
change_treatment = post_treatment - pre_treatment
change_control = post_control - pre_control
```

```python
# check normality
from scipy import stats

stat_t, p_t = stats.shapiro(change_treatment)
stat_c, p_c = stats.shapiro(change_control)
print(f'Treatment change p={p_t:.3f}, Control change p={p_c:.3f}')
```

### **Difference-in-Differences(t-test on change scores)**

```python
t_stat, p_val = stats.ttest_ind(change_treatment, change_control, alternative='greater')
print(f'DiD t-test: t = {t_stat:.3f}, p-one-tailed = {p_val:.3f}')
```

p < 0.05, reject H0 and conclude "revenue increase in treatment beyond the control group"

```python
# effect size
pooled_sd = np.sqrt(((n_per_group-1)*change_treatment.std(ddof=1)**2 +
                     (n_per_group-1)*change_control.std(ddof=1)**2) / (2*n_per_group - 2))
cohens_d = (change_treatment.mean() - change_control.mean()) / pooled_sd
print(f"Cohen's d = {cohens_d:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter([0]*n_per_group, change_control, color='grey', label='Control change')
plt.scatter([1]*n_per_group, change_treatment, color='green', label='Treatment change')
plt.xticks([0,1], ['Control', 'Treatment'])
plt.ylabel('Revenue Change')
plt.title('Difference-in-Differences: Pre-Post Revenue Change')
plt.legend()
plt.show()
```

From the plot above, we can see that treatment group shows larger increase than control

**Conclusion:**
Treatment revenue increased significantly more than control (DiD t-test). Effect is large and practically meaningful (Cohen's d > 1). Sample size planning ensures study is powered (80%) to detect the expected effect. Random assignment + control group results in strong causal inference: **"advertising caused additional revenue increase beyond general trends."**

## **Non-Parametric Tests**

Why do we need these? If normality assumptions are violated or sample size is small, we can use non-parametric alternatives. Most common non-parametric tests are designed to test the median (or ranks) rather than the mean. Mean-Whitney U is sometimes described as testing "difference in distributions", not strictly median. But for many practical cases, if the shapes of the distributions are similar, a significant Mann-Whitney U result implies difference in medians.

Medians are robust to outliers and skewed data. Ranks focus on order not exact values, which avoids relying on the assumption about the mean. For small or non-normal samples, this gives more reliable inerence.

Hence, non-paramteric tests are robut alternatives to t-tests and ANOVA when assumptions are violated.

```python
np.random.seed(42)

# 1. Observational Study

print("\n=== Observational Study ===")

n_sellers = 30

# Pre and Post revenue for single group (paired design)
pre_revenue = np.random.normal(70, 8, n_sellers)
post_revenue = pre_revenue + np.random.normal(5, 5, n_sellers)

obs_df = pd.DataFrame({'Pre': pre_revenue, 'Post': post_revenue})

# One-Sample Case
# Test if pre-revenue differs from target 70
stat, p = stats.wilcoxon(pre_revenue - 70)
print(f"Observational One-Sample (Wilcoxon) test: stat={stat:.3f}, p={p:.3f}")

# Paired Case
stat, p = stats.wilcoxon(obs_df['Post'], obs_df['Pre'])
print(f"Observational Paired (Wilcoxon) test: stat={stat:.3f}, p={p:.3f}")

# Visualize Pre vs Post
sns.boxplot(data=obs_df)
plt.title("Observational Study: Pre vs Post Revenue")
plt.show()
```

```python
# Observational Difference-in-Differences (DiD)

# Suppose two natural groups (treatment and control)
n_per_group = 25
pre_treatment = np.random.normal(70, 8, n_per_group)
post_treatment = pre_treatment + np.random.normal(5, 5, n_per_group)  # Treatment effect

pre_control = np.random.normal(70, 8, n_per_group)
post_control = pre_control + np.random.normal(1, 5, n_per_group)      # General trend

# Change scores
change_treatment = post_treatment - pre_treatment
change_control = post_control - pre_control

# Non-parametric DiD test
stat, p = stats.mannwhitneyu(change_treatment, change_control, alternative='greater')
print(f"Observational DiD (Mann-Whitney U) test: stat={stat:.3f}, p={p:.3f}")

plt.scatter([0]*n_per_group, change_control, color='grey', label='Control change')
plt.scatter([1]*n_per_group, change_treatment, color='green', label='Treatment change')
plt.xticks([0,1], ['Control', 'Treatment'])
plt.ylabel('Revenue Change')
plt.title('Observational DiD: Pre-Post Revenue Change')
plt.legend()
plt.show()
```

```python
# Difference-in-Differences (experimental)

# Randomized treatment and control
pre_treat = np.random.normal(70, 8, n_per_group)
post_treat = pre_treat + np.random.normal(5, 5, n_per_group)  # treatment effect

pre_ctrl = np.random.normal(70, 8, n_per_group)
post_ctrl = pre_ctrl + np.random.normal(1, 5, n_per_group)    # general trend

change_treat = post_treat - pre_treat
change_ctrl = post_ctrl - pre_ctrl

# Non-parametric DiD test
stat, p = stats.mannwhitneyu(change_treat, change_ctrl, alternative='greater')
print(f"Experimental DiD (Mann-Whitney U) test: stat={stat:.3f}, p={p:.3f}")

plt.scatter([0]*n_per_group, change_ctrl, color='grey', label='Control change')
plt.scatter([1]*n_per_group, change_treat, color='blue', label='Treatment change')
plt.xticks([0,1], ['Control', 'Treatment'])
plt.ylabel('Revenue Change')
plt.title('Experimental DiD: Pre-Post Revenue Change')
plt.legend()
plt.show()
```

## **Bootstrapping Method**

Bootstrapping is a resampling-based method that allows us to estimate sampling distributions, confidence intervals, and p-values without strict parametric assumptions (like normality).
It is especially useful for:
- Small samples
- Non-normal data
- Complex statistics (e.g., medians, differences in means)

The key idea is:

1. Take B resamples from the dataset with replacement.

2. Compute the statistic of interest (mean, difference, etc.) for each resample.

3. Use the distribution of the resampled statistics to estimate confidence intervals or p-values.


Bootstrapping can be applied to:
- One-sample:	Mean (or median) vs target
- Paired:	Mean (or median) difference Pre-Post
- Independent:	Difference in group means
- Difference-in-Differences:	Difference of change scores between groups

```python
def bootstrap_statistic(data1, data2=None, stat_func=np.mean, n_boot=10000):
  """Compute bootstrap distribution of statistic.
  If data2 is None, do one-sample or paired differences.
  If data2 is provided, compute difference in statistic between data1 and data2.
  """
  boot_stats = []
  n1 = len(data1)
  if data2 is None:
    # One-sample or paired differences
    for _ in range(n_boot):
      sample = np.random.choice(data1, size=n1, replace=True)
      boot_stats.append(stat_func(sample))
  else:
    n2 = len(data2)
    for _ in range(n_boot):
      sample1 = np.random.choice(data1, size=n1, replace=True)
      sample2 = np.random.choice(data2, size=n2, replace=True)
      boot_stats.append(stat_func(sample1) - stat_func(sample2))
  return np.array(boot_stats)
```

```python
np.random.seed(42)

# 1. Observational Study

print("\n=== Observational Study ===")

n_sellers = 30
pre = np.random.normal(70, 8, n_sellers)
post = pre + np.random.normal(5, 5, n_sellers)

# One-Sample vs target (70)
diff_from_target = pre - 70
boot_one_sample = bootstrap_statistic(diff_from_target)
ci_one_sample = np.percentile(boot_one_sample, [2.5, 97.5])
print(f"Observational One-Sample Bootstrapped 95% CI: {ci_one_sample}")

# Paired Pre-Post
change = post - pre
boot_paired = bootstrap_statistic(change)
ci_paired = np.percentile(boot_paired, [2.5, 97.5])
print(f"Observational Paired Bootstrapped 95% CI: {ci_paired}")

# Difference-in-Differences (observational)
n_group = 25
pre_treat = np.random.normal(70, 8, n_group)
post_treat = pre_treat + np.random.normal(5, 5, n_group)
pre_ctrl = np.random.normal(70, 8, n_group)
post_ctrl = pre_ctrl + np.random.normal(1, 5, n_group)

change_treat = post_treat - pre_treat
change_ctrl = post_ctrl - pre_ctrl

boot_did_obs = bootstrap_statistic(change_treat, change_ctrl)
ci_did_obs = np.percentile(boot_did_obs, [2.5, 97.5])
print(f"Observational DiD Bootstrapped 95% CI: {ci_did_obs}")
```

```python
# 2. Experimental Study

print("\n=== Experimental Study ===")

# Independent groups
treatment = np.random.normal(75, 8, n_group)
control = np.random.normal(70, 8, n_group)

boot_ind = bootstrap_statistic(treatment, control)
ci_ind = np.percentile(boot_ind, [2.5, 97.5])
print(f"Experimental Independent Bootstrapped 95% CI: {ci_ind}")

# Paired Pre-Post
pre_exp = np.random.normal(70, 8, n_group)
post_exp = pre_exp + np.random.normal(5, 5, n_group)
change_exp = post_exp - pre_exp

boot_paired_exp = bootstrap_statistic(change_exp)
ci_paired_exp = np.percentile(boot_paired_exp, [2.5, 97.5])
print(f"Experimental Paired Bootstrapped 95% CI: {ci_paired_exp}")

# Difference-in-Differences (experimental)
pre_treat_exp = np.random.normal(70, 8, n_group)
post_treat_exp = pre_treat_exp + np.random.normal(5, 5, n_group)
pre_ctrl_exp = np.random.normal(70, 8, n_group)
post_ctrl_exp = pre_ctrl_exp + np.random.normal(1, 5, n_group)

change_treat_exp = post_treat_exp - pre_treat_exp
change_ctrl_exp = post_ctrl_exp - pre_ctrl_exp

boot_did_exp = bootstrap_statistic(change_treat_exp, change_ctrl_exp)
ci_did_exp = np.percentile(boot_did_exp, [2.5, 97.5])
print(f"Experimental DiD Bootstrapped 95% CI: {ci_did_exp}")

plt.hist(boot_did_exp, bins=50, alpha=0.7)
plt.axvline(ci_did_exp[0], color='red', linestyle='--')
plt.axvline(ci_did_exp[1], color='red', linestyle='--')
plt.title("Bootstrapped DiD Distribution (Experimental)")
plt.xlabel("Difference in change")
plt.ylabel("Frequency")
plt.show()
```

## **ANOVA**

### **Observational Study**

Scenario: Suppose we have sellers' revenue data collected for different ad strategies, but we did not assign them. We have 3 different types of ANOVA:

- Can be one-way (single factor, e.g., Strategy A/B/C)
- Can be two-way (e.g., Strategy A/B/C x Region East/West)
- Can be repeated measures (same sellers under multiple strategies, pre/post design)

**Hypothesis Formulation**

**One way ANOVA:** Used when we want to compare the mean of one numerical variable (revenue) across three or more independent groups (three ad strategies A,B,C)

$H_0: \mu_A = \mu_B = \mu_C$ (all means equal)

$H_1$: At least one mean differs

**Two way ANOVA:** Analyzes how two different categorical factors influence numerical variable.

Factor 1: Strategy (A, B, C)

Factor 2: Region (East, West, North, South)

Two way anova lets us test multiple effects at the same time: Main effect of Strategy, Main effect of Region, Interaction effect between Strategy and Region.

Main effects + interaction:

- $H_0 strategy: \mu_{strategyA} = \mu_{strategyB} = \mu_{strategyC}$,  Average revenue does not differ across strategies when ignoring region

- $H_0 region:\mu_{East} = \mu_{West} = \mu_{North} = \mu_{South}$,  Sales region does not influence mean revenue

- $H_0 interaction:$ The effect of strategy is same across all regions

**Repeated Measures (Paired ANOVA)**

Compare same sellers revenue under multiple strategies (within-subject)

$H_0: \mu_A = \mu_B = \mu_C$ Mean revenue is equal across strategies

$H_1:$ At least one mean differs.

**Assumptions:** Independence can be violated if sellers self-select strategies (observational study). Data is normally distributed per group (Shapiro-Wilk test per group). All groups have equal variances (Levene's test). Sphericity is only for repeated measures ANOVA (Mauchy test). Sphericity means the differences between all pairs of repeated conditions have equal variances.

```python
# observational study dataset (sellers self-selected ad strategies A, B, C)

np.random.seed(42)

strategy_A = np.random.normal(70, 8, 12) # 12 sellers chose A
strategy_B = np.random.normal(75, 10, 15) # 15 sellers chose B
strategy_C = np.random.normal(73, 9, 10) # 10 sellers chose C

obs_df = pd.DataFrame({'Revenue': np.concatenate([strategy_A, strategy_B, strategy_C]),
                       'Strategy': ['A']*len(strategy_A) + ['B']*len(strategy_B) + ['C']*len(strategy_C)})

obs_df.head()
```

#### **One Way ANOVA**

```python
# One Way ANOVA

# check independence
from scipy.stats import shapiro

for name, group in zip(['A','B','C'], [strategy_A, strategy_B, strategy_C]):
  stat, p = shapiro(group)
  if p < 0.05:
    print(f"{name}: Shapiro-Wilk p={p:.3f} → normality violated")
  else:
    print(f"{name}: Shapiro-Wilk p={p:.3f} → normality assumption OK")
```

```python
# check variance (equal variances homoscedasticity)

from scipy.stats import levene

stat, p = levene(strategy_A, strategy_B, strategy_C)
if p < 0.05:
  print(f"Levene test p={p:.3f} → variance homoscedastic assumption violated")
else:
  print(f"Levene test p={p:.3f} → variance homoscedastic assumption OK")

# this is important because unequal variaces plus unqual sample sizes can bias ANOVA
```

Observational data may violate independence. Some sellers may influence each other's choice of strategy.

```python
# ANOVA test

from scipy.stats import f_oneway

f_stat, p_val = f_oneway(strategy_A, strategy_B, strategy_C)
print(f"ANOVA: F={f_stat:.3f}, p={p_val:.3f}")
if p_val < 0.05:
  print("Significant difference exists between groups")
else:
  print("No significant difference")
```

No significant difference in mean revenue between strategies.

```python
# effect size

all_data = np.concatenate([strategy_A, strategy_B, strategy_C])
grand_mean = all_data.mean()
ss_between = sum(len(g)*(g.mean()-grand_mean)**2 for g in [strategy_A, strategy_B, strategy_C])
ss_total = sum((x-grand_mean)**2 for x in all_data)
eta_sq = ss_between / ss_total
print(f"Eta squared = {eta_sq:.3f}")
```

4.1% of variance in revenue is explained by strategy which is very small effect.

```python
# Post-Hoc Test (Tukey HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd
df = pd.DataFrame({
    'Revenue': np.concatenate([strategy_A, strategy_B, strategy_C]),
    'Strategy': ['A']*len(strategy_A) + ['B']*len(strategy_B) + ['C']*len(strategy_C)
})

tukey = pairwise_tukeyhsd(endog=df['Revenue'], groups=df['Strategy'], alpha=0.05)
print(tukey)
```

No pairwise comparisions are significant. Since no significant overall effect, post-hoc is not strictly necessary but good practice to report anyway.

#### **non-parametric alternative**

```python
# non-parametric alternative if assumptions fail

from scipy.stats import kruskal

stat, p = kruskal(strategy_A, strategy_B, strategy_C)
if p < 0.05:
  print("Significant difference in medians (Kruskal-Wallis)")
else:
  print("No significant difference (Kruskal-Wallis)")
```

Non parametric test confirms no significant difference in medians.

```python
# bootstrapping but optional
n_boot = 10000
boot_diff = []

for _ in range(n_boot):
  boot_samples = [np.random.choice(g, len(g), replace=True) for g in [strategy_A, strategy_B, strategy_C]]
  boot_diff.append(np.max([s.mean() for s in boot_samples]) - np.min([s.mean() for s in boot_samples]))

ci_lower = np.percentile(boot_diff, 2.5)
ci_upper = np.percentile(boot_diff, 97.5)
print(f"Bootstrapped 95% CI for max-min mean difference: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

#### **Two Way ANOVA with Unequal Group Sizes and Variance Violation**

#### **one way-experimental ANOVA**

**Sample size / power for a one way-experimental ANOVA**

Goal: detect a medium effect (cohen's $f=0.25$, $\alpha = 0.05$, $power=0.8$, with 3 groups.

Notes:
- f = 0.10 small effect
- f = 0.25 medium effect
- f = 0.40 large effect

```python
# sample size calculation (one-way ANOVA) using statsmodels
from statsmodels.stats.power import FTestAnovaPower
import math

analysis = FTestAnovaPower()
alpha = 0.05
power_target = 0.8
effect_f = 0.25   # Cohen's f: 0.10 small, 0.25 medium, 0.40 large
k_groups = 3

# solve for total sample size required
n_total_needed = analysis.solve_power(effect_size=effect_f, nobs=None, alpha=alpha, power=power_target, k_groups=k_groups)
n_per_group_needed = math.ceil(n_total_needed / k_groups)
print(n_total_needed, n_per_group_needed)
```

To reliably detect a medium difference between three strategies with acceptable power, we want about 53 sellers per group.

```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Total 30 sellers, randomly assigned to 3 strategies equally
n_sellers = 30
strategies = ['A', 'B', 'C']
n_per_group = n_sellers // 3

# Random assignment
np.random.shuffle(strategies * n_per_group)  # pseudo-random, balanced

# Generate revenue assuming some treatment effect
revenue_A = np.random.normal(70, 8, n_per_group)      # control baseline
revenue_B = np.random.normal(75, 8, n_per_group)      # treatment effect
revenue_C = np.random.normal(72, 8, n_per_group)      # another treatment effect

exp_df = pd.DataFrame({
    'Revenue': np.concatenate([revenue_A, revenue_B, revenue_C]),
    'Strategy': ['A']*n_per_group + ['B']*n_per_group + ['C']*n_per_group
})
exp_df.head()
```

Random assignment ensures groups are balanced which reduces confounding. Mean differences can now be interpreted causally. This contrasts with observational ANOVA where groups were self-directed.

**Power with current sample (30 total, 10 per group)**

```python
# compute achieved power with your sample size
n_total_user = 30   # 10 per group
power_achieved = analysis.solve_power(effect_size=effect_f, nobs=n_total_user, alpha=alpha, power=None, k_groups=k_groups)
print(power_achieved)
```

With only 10 sellers per group, we have very low power to detect medium effect. It means that we have around 80% chance of missing a true medium effect (high type II error). We have to increase the sampple size to 53 per group if we want 80% power to detect a medium effect.

If adding subjects is impossible then we need to consider within-subject (repeated measures) design which is much more powerful and needs very few subjects. ## **Chi-Squared Test**

Assumptions: Chi-squared requires counts, not percentages. Categories must be mutually exclusive and expected cell frequency should be greater than five for most cells. If this is violated then we use Fisher's Exact Test or Monte Carlo Simulation

### **Observational Study**


#### **Chi-squared Test of Independence**

Context: Sellers have chosen strategies A, B, C on their own. We want to see if strategy is associated with being High Revenue.

$H_0$: Strategy and High Revenue are independent (no association)

$H_1$: Strategy and High Revenue are associated (no independent)

Note: Even if we find association, we cannot claim strategy causes high revenue because sellers self-select strategies.

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Observational: sellers self-select strategies
strategy_A = np.random.normal(70, 8, 12)
strategy_B = np.random.normal(75, 10, 15)
strategy_C = np.random.normal(73, 9, 10)

obs_df = pd.DataFrame({
    'Revenue': np.concatenate([strategy_A, strategy_B, strategy_C]),
    'Strategy': ['A']*12 + ['B']*15 + ['C']*10
})

# Categorize revenue as High/Low
median_rev = np.median(obs_df['Revenue'])
obs_df['RevenueCategory'] = np.where(obs_df['Revenue']>median_rev, 'High', 'Low')
```

```python
# creating high/low revenue categories based on median

median_rev = obs2_df['Revenue'].median()

obs2_df['HighRevenue'] = np.where(obs2_df['Revenue'] > median_rev, 'High', 'Low')

contingency = pd.crosstab(obs2_df['Strategy'], obs2_df['HighRevenue'])
contingency
```

The above table shows the observed counts, what we actually saw in the data.

```python
# check expected cell counts (assumption for chi-square)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency, correction=False)
expected_df = pd.DataFrame(expected, index=contingency.index, columns=contingency.columns)
print("Expected counts:\n", expected_df)
```

Chi-square assumes expected counts in each cell are typically ≥ 5. We print expected counts to check this. If many expected counts < 5, use Fisher’s exact (only for 2×2) or Monte-Carlo / permutation methods.

From the results above, the assumption is satisfied suggesting little deviation.

```python
# Run chi-square (Pearson) test
chi2, p, dof, expected = chi2_contingency(contingency, correction=False)
print(f"Chi2 = {chi2:.3f}, p = {p:.3f}, dof = {dof}")
```

p value of 0.976 suggests there is no evidence of association. The observed pattern high/low revenue across strategies is almost exactly what we would expect by random chanve.

We fail to reject the null and conclude strategy choice is not statistically associated with revenue category in this observational dataset.

```python
# effect size (Cramer's V)

import numpy as np

def cramers_v(chi2, n, r, k):
  return np.sqrt(chi2 / (n * (min(r, k) - 1)))

chi2, p, dof, expected = chi2_contingency(contingency, correction=False)
n = contingency.values.sum()
r, k = contingency.shape
cramers = cramers_v(chi2, n, r, k)
print(f"Cramer's V = {cramers:.3f}")
```

Cramér’s V measures the strength of association (0 = none, 1 = perfect). Rough interpretation (common rule of thumb): for df=1, 0.1 small, 0.3 medium, 0.5 large; for larger tables, interpret conservatively.

From the results above, we can conclude that even if the sample size increases, the association would likely remain tiny.

```python
# post hoc / pairwise comparisons (if overall is significant)

# If overall significant, find which strategies differ in high proportions. Two common approaches

# Pairwise chi-square or Fisher with Bonferroni correction
from itertools import combinations
from scipy.stats import chi2_contingency, fisher_exact

pairs = list(combinations(contingency.index, 2))
results = []
for a, b in pairs:
  sub = obs2_df[obs2_df['Strategy'].isin([a, b])]
  table = pd.crosstab(sub['Strategy'], sub['HighRevenue'])
  # if 2x2 and small expected counts, use fisher_exact
  if table.shape == (2,2) and (table.values < 5).sum() > 0:
    od, pval = fisher_exact(table)
    method = 'fisher'
  else:
    chi2_p = chi2_contingency(table, correction=False)[1]
    pval = chi2_p
    method = 'chi2'
  results.append((a, b, pval, method))

# Bonferroni adjust
m = len(results)
results_adj = [(a,b,pval,method,min(1,pval*m)) for (a,b,pval,method) in results]
print("pairwise (Bonferroni adjusted p):", results_adj)
```

Compares A vs B, A vs C and B vs C. Apply bonferroni correction (multiply p by number of tests) to control family-wise error.

All p values are extremely non significant. No pair of strategies differs in the proportion of high revenue sellers.

```python
# Adjusted standardized residuals (to see which cells contribute the most)

import numpy as np
obs = contingency.values
exp = expected
std_resid = (obs - exp) / np.sqrt(exp)  # approximate
std_resid_df = pd.DataFrame(std_resid, index=contingency.index, columns=contingency.columns)
print("Standardized residuals:\n", std_resid_df)
```

## **Advanced Topics in Observational Study**

### **Observational Studies (Self-Selected Marketing Strategies)**

Scenario:

In observational studies, sellers choose their own marketing strategies (A, B, or C). This self-selection can introduce confounding, because the choice of strategy may depend on other factors such as: Seller experience, Marketing budget, Region or other unobserved characteristics. Because of this confounding, classical statistical tests like t-tests, ANOVA, or chi-squared tests cannot establish causality

#### **Regression Adjustment**

To account for observed confounders, we can use regression adjustment. Here, we model the probability of a seller achieving high revenue as a function of: Marketing strategy, Experience and Marketing budget. This helps us estimate the effect of each strategy while controlling for other factors.

```python
# Example: Does Strategy B Increase High Revenue Probability?

import pandas as pd
import numpy as np
import statsmodels.api as sm

np.random.seed(42)
n_sellers = 150

obs_df = pd.DataFrame({
    'Seller': range(n_sellers),
    'Experience': np.random.randint(1, 11, n_sellers),
    'MarketingBudget': np.random.normal(1000, 200, n_sellers),
    'Strategy': np.random.choice(['A', 'B', 'C'], n_sellers, p=[0.4, 0.3, 0.3])
})

strategy_effect = {'A': 5, 'B': 7, 'C': 6}
obs_df['Revenue'] = (obs_df['Experience']*2 +
                     obs_df['MarketingBudget']*0.01 +
                     obs_df['Strategy'].map(strategy_effect) +
                     np.random.normal(0, 5, n_sellers))

obs_df['HighRevenue'] = (obs_df['Revenue'] > obs_df['Revenue'].median()).astype(int)
```

```python
obs_df.head()
```

```python
# Regression adjustment
X = pd.get_dummies(obs_df[['Strategy','Experience','MarketingBudget']], drop_first=True).astype(float)
X = sm.add_constant(X)
y = obs_df['HighRevenue'].astype(float)

logit_model = sm.Logit(y, X).fit()
print(logit_model.summary())
```

Above, we analyzed how different marketing strategies (A, B, C) influence the probability of achieving high revenue, while controlling for seller experience and marketing budget using logistic regression.

Experience and budget clearly influence revenue outcomes, but in self-selected marketing strategies, the impact of Strategy B or C is not statistically clear after adjusting for these factors. Observational data alone cannot confirm that any strategy causes higher revenue.

In other words, since this is an observational study, strategy choice may be influenced by other factors not included in the model (unobserved confounding), so these results do not imply causality.

#### **Propensity Score Methods**

**Background:** In observational studies, sellers choose their own strategies. This can bias simple comparisons because factors like experience and marketing budget influence both strategy choice and revenue.

Propensity score methods help reduce this bias by:
- Estimating the probability (propensity) of choosing a strategy given observed covariates.
- Matching sellers with similar propensity scores from treated (Strategy A) and control (Strategies B or C) groups.
- Comparing outcomes in these matched groups to estimate a more causal effect.

**Implementation:** We estimated the effect of Strategy A compared to other strategies, adjusting for Experience and Marketing Budget

**Steps:**
- Created a treatment variable: 1 if Strategy A, 0 otherwise.
- Standardized covariates and fitted a logistic regression to predict the probability of choosing Strategy A (the propensity score).
- Used nearest-neighbor matching to pair each Strategy A seller with a non-A seller who had a similar propensity score.
- Compared average revenue between treated and matched control sellers.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

obs_df['TreatmentA'] = (obs_df['Strategy']=='A').astype(int)
covariates = ['Experience','MarketingBudget']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(obs_df[covariates])
y_treat = obs_df['TreatmentA']

ps_model = LogisticRegression().fit(X_scaled, y_treat)
obs_df['PropensityScore'] = ps_model.predict_proba(X_scaled)[:,1]

treated = obs_df[obs_df['TreatmentA']==1]
control = obs_df[obs_df['TreatmentA']==0]

nn = NearestNeighbors(n_neighbors=1).fit(control[['PropensityScore']])
distances, indices = nn.kneighbors(treated[['PropensityScore']])
matched_control = control.iloc[indices.flatten()]

treated_mean = treated['Revenue'].mean()
control_mean = matched_control['Revenue'].mean()
print(f"Estimated causal effect of Strategy A: {treated_mean - control_mean:.2f}")
```

After adjusting for experience and marketing budget, sellers using Strategy A earn on average $4.58 less than similar sellers using Strategies B or C. Strategy A does not appear to improve revenue relative to other strategies; in fact, it shows a small negative effect.

#### **Inverse Probability Weighting (IPW)**

Inverse Probability Weighting (IPW) is a method to adjust for confounding by:
- Weighting each seller by the inverse of the probability of receiving the treatment they actually received (propensity score).
- Creating a pseudo-population where covariates (experience, marketing budget) are balanced across treatment groups.
- Comparing outcomes in this weighted population to estimate the causal effect.

**Implementation**

We estimated the effect of Strategy A as follows:

- Compute weights for each seller
- Calculate weighted average revenue for Strategy A and non-A sellers

```python
obs_df['Weight'] = np.where(
    obs_df['TreatmentA']==1,
    1/obs_df['PropensityScore'],
    1/(1-obs_df['PropensityScore'])
)
```

```python
weighted_mean_A = np.average(
    obs_df.loc[obs_df['TreatmentA']==1, 'Revenue'],
    weights=obs_df.loc[obs_df['TreatmentA']==1, 'Weight']
)
weighted_mean_control = np.average(
    obs_df.loc[obs_df['TreatmentA']==0, 'Revenue'],
    weights=obs_df.loc[obs_df['TreatmentA']==0, 'Weight']
)
print(weighted_mean_A, weighted_mean_control)
```

After accounting for experience and marketing budget using IPW, Strategy A slightly reduces revenue compared to other strategies. This aligns with the earlier propensity score matching result, suggesting that Strategy A does not outperform other strategies

#### **Instrumental Variables (IV): Accounting for Unobserved Confounding**

**Background:** In observational studies, self-selection of marketing strategies can bias estimates due to unobserved confounders (factors affecting both strategy choice and revenue that we cannot measure).Instrumental Variables (IV) help address this by using a variable (instrument) that:

- Affects strategy choice (the treatment)
- Does not directly affect revenue (the outcome) except through the strategy

In our example:

- Instrument: Region (East vs West)
- Treatment: Marketing strategy (A, B, or C, coded numerically)
- Outcome: Revenue

Using IV allows us to estimate the causal effect of strategy on revenue, even with unobserved confounding.

**Implementation**

We use two-stage least squares (2SLS):

- Stage 1: Predict strategy choice using the instrument (region).
- Stage 2: Regress revenue on the predicted strategy values from Stage 1.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.sandbox.regression.gmm import IV2SLS

np.random.seed(42)
n_sellers = 150

# Simulate data
obs_df = pd.DataFrame({
    'Revenue': np.random.normal(50, 10, n_sellers),
    'Strategy': np.random.choice(['A','B','C'], n_sellers),
    'Region': np.random.choice(['East','West'], n_sellers)
})

# Encode strategy as numeric
obs_df['StrategyCode'] = obs_df['Strategy'].map({'A':0,'B':1,'C':2}).astype(float)

# Encode instrument (Region) as numeric dummies
Z = pd.get_dummies(obs_df['Region'], drop_first=True).astype(float)

# Add constant to instrument matrix
Z = sm.add_constant(Z)

# Outcome
Y = obs_df['Revenue'].astype(float)

# Endogenous variable
T = obs_df['StrategyCode']

# Fit IV regression
iv_model = IV2SLS(endog=Y, exog=T, instrument=Z).fit()
print(iv_model.summary())
```

We investigated how different marketing strategies (A, B, C) affect seller revenue using four methods: regression adjustment, propensity score matching, inverse probability weighting (IPW), and instrumental variables (IV). Using regression adjustment to control for experience and marketing budget, we found that Strategy B and C did not significantly increase the probability of achieving high revenue compared to Strategy A, highlighting that experience and budget are the strongest predictors. Propensity score matching revealed that, after balancing for observed covariates, Strategy A sellers earned on average 4.58 less than similar non-A sellers, and IPW produced a similar result, with Strategy A slightly underperforming other strategies by 1.63. However, when using IV analysis with region as an instrument to account for unobserved confounding, we found a strong positive causal effect: moving from one strategy to a higher-coded strategy (A to B to C) increased revenue by approximately 50.8 units. Overall, these results suggest that while observational adjustments indicate Strategy A is not particularly effective, accounting for hidden confounders reveals that upgrading to higher strategies can substantially increase revenue. Experience and marketing budget remain important, but causal analysis highlights the significant benefits of more advanced marketing strategies.

#### **Sensitivity Analysis**

**How robust are our results?**

**Background:** Even after adjusting for observed factors, there may still be unmeasured confounders i.e., hidden factors that affect both strategy choice and revenue.

Sensitivity analysis asks: How strong would an unmeasured confounder have to be to completely explain away the observed effect?

This is often expressed as an E-value. A higher E-value means the observed effect is more robust to hidden bias.

```python
# Suppose we observed an odds ratio (OR) of 1.5 for a strategy increasing high revenue:

OR_observed = 1.5
e_value = OR_observed + np.sqrt(OR_observed * (OR_observed - 1))
e_value

# OR_observed: The observed effect of the strategy (1.5 means 50% higher odds of high revenue)

# E-value formula: Measures the minimum strength of an unmeasured confounder needed to reduce the
# observed effect to no effect (OR=1)
```

A hidden confounder would need to increase both the odds of receiving the strategy and the odds of achieving high revenue by at least 2.37 times to completely explain away the observed effect.Since this is relatively high, the observed effect is moderately robust, it is unlikely that a hidden confounder alone would nullify the finding.

Think of the E-value as a “stress test” for our result. The higher the E-value, the stronger a confounder would need to be to overturn our observed effect. Here, an E-value of 2.37 suggests the observed effect of the strategy is fairly credible even in the presence of unmeasured confounding.

#### **Difference-in-Difference (DiD)**

**Evaluating the effect of Marketing Intervention**

Background: Difference-in-Differences (DiD) is used to estimate the causal effect of an intervention by comparing changes over time between a treated group and a control group.
- Treated group: Sellers who adopted a new strategy (Treatment)
- Control group: Sellers who did not change strategy (Control)
- Pre-period: Before intervention
- Post-period: After intervention

DiD calculates the difference in revenue changes between the two groups, isolating the effect of the strategy from general time trends.

**Implementation**

We simulated revenue for 100 sellers across two periods (Before and After) with the following logic:

All sellers start with a baseline revenue (~50 units).

- Post-period effect: Revenue increases by 5 units for everyone due to time trends.

- Treatment effect: Treated sellers in the post-period get an additional 10 units.

We then defined:

- Post: 1 if after intervention, 0 otherwise

- Treat: 1 if in treatment group, 0 if control

DiD: Interaction of Post and Treat (captures the treatment effect)

Finally, we ran a linear regression:

$$Revenue=const+β_1Post+β_2Treat+β_3(Post \times Treat)+ϵ$$

**Result Interpretation:**
- Post Coefficient($β_1$): Captures the general revenue change over time for all sellers (time trend)
- Treatment Coefficient($β_2$): Captures baseline difference between treatment and control groups before the intervention
- DiD Coefficient($\beta_3$): Captures the causal effect of the intervention, the extra revenue due to the treatment strategy

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

np.random.seed(42)
n_sellers = 100

# Create the panel dataset
df = pd.DataFrame({
    'Seller': np.repeat(range(n_sellers), 2),
    'Period': list(np.tile(['Before', 'After'], n_sellers))
})

# Assign strategy to each seller (length = n_sellers) and repeat for both periods
strategy = np.random.choice(['Control', 'Treatment'], n_sellers)
df['Strategy'] = np.repeat(strategy, 2)

# Simulate revenue
df['Revenue'] = (
    50
    + (df['Period'] == 'After') * 5  # time effect
    + ((df['Strategy'] == 'Treatment') & (df['Period'] == 'After')) * 10  # DiD effect
    + np.random.normal(0, 3, 2 * n_sellers)
)

# DiD variables
df['Post'] = (df['Period'] == 'After').astype(int)
df['Treat'] = (df['Strategy'] == 'Treatment').astype(int)
df['DiD'] = df['Post'] * df['Treat']

# Run DiD regression
model = sm.OLS(df['Revenue'], sm.add_constant(df[['Post', 'Treat', 'DiD']])).fit()
print(model.summary())
```

The Difference-in-Differences analysis estimated the causal effect of a marketing strategy by comparing revenue changes over time between treated and control sellers. The Treat coefficient was not significant, indicating that the treated and control groups had similar revenue before the intervention, which supports the DiD assumption that they would have followed parallel trends in the absence of the strategy. The Post coefficient showed a general revenue increase of about 5.6 units over time for all sellers. Importantly, the DiD (interaction) coefficient was approximately 10 units and highly significant, meaning that treated sellers earned around 10 units more than controls due to the strategy. Overall, these results indicate that the marketing strategy had a positive causal effect on revenue, beyond general time trends

### **Experimental Studies (Randomized Assignment)**

**Background:**

In experimental studies, sellers are randomly assigned to marketing strategies (A, B, or C). #### **ANCOVA (Analysis of Covariance)**

**Effect of Randomly Assigned Marketing Strategies**

ANCOVA (Analysis of Covariance) is a method that combines ANOVA (comparing group means) with regression adjustment for covariates. Here, it allows us to:

- Estimate the causal effect of strategy on revenue.
- Adjust for Experience, which influences revenue.
- Increase the precision of the estimated strategy effect.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

np.random.seed(42)
n_sellers = 150

# Create experimental dataset
exp_df = pd.DataFrame({
    'Seller': range(n_sellers),
    'Experience': np.random.randint(1, 11, n_sellers),
    'Strategy': np.repeat(['A','B','C'], n_sellers//3)
})

# Define strategy effects
strategy_effect = {'A': 50, 'B': 55, 'C': 53}

# Generate revenue with some random noise
exp_df['Revenue'] = (
    exp_df['Experience']*2
    + exp_df['Strategy'].map(strategy_effect)
    + np.random.normal(0, 5, n_sellers)
)

# ANCOVA: Revenue ~ Strategy + Experience
model = smf.ols('Revenue ~ Strategy + Experience', data=exp_df).fit()
print(model.summary())
```

In this randomized experiment, sellers were assigned to Marketing Strategies A, B, or C, and revenue was analyzed while adjusting for experience to improve precision. The results show that Strategy B significantly increased revenue by about 5.35 units compared to Strategy A, while Strategy C showed a smaller, non-significant increase of 1.72 units. Experience had a strong positive effect, with each additional year increasing revenue by approximately 2.18 units. The model explains about 64% of the variation in revenue, indicating a good fit. Because the strategies were randomly assigned, the observed differences can be interpreted as causal effects, and adjusting for experience allows for a more precise estimate of the strategy effect. Overall, these results suggest that Strategy B is the most effective in increasing revenue, while experience also plays an important role

#### **Mixed-Effects / Multilevel Models**

In real-world data, sellers may be grouped into clusters such as regions, and revenue can vary both because of marketing strategies and because of regional differences. Ignoring this nested structure can underestimate variability and produce misleading inferences.

Mixed-effects models (or multilevel models) handle this by:

- Modeling fixed effects for predictors like Strategy and Experience.
- Modeling random effects for clusters (e.g., Region) to capture variation in baseline revenue across groups.
- Accounting for the nested structure of the data, producing more accurate standard errors and estimates.

This is particularly useful when there are multiple regions so that random effects are identifiable.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

np.random.seed(42)
n_sellers = 300
n_regions = 20

# -----------------------------
# Create dataset
# -----------------------------
exp_df = pd.DataFrame({
    'Seller': range(n_sellers),
    'Experience': np.random.randint(1, 11, n_sellers),
    'Strategy': np.repeat(['A','B','C'], n_sellers//3)
})

# Assign sellers randomly to regions
regions = [f'Region_{i+1}' for i in range(n_regions)]
exp_df['Region'] = np.random.choice(regions, n_sellers)

# Define strategy effects
strategy_effect = {'A': 50, 'B': 55, 'C': 53}

# Create random intercepts with **larger variance**
region_effects = dict(zip(regions, np.random.normal(0, 10, n_regions)))  # larger between-region variation

# Generate revenue with smaller residual noise
exp_df['Revenue'] = (
    exp_df['Experience']*2
    + exp_df['Strategy'].map(strategy_effect)
    + exp_df['Region'].map(region_effects)
    + np.random.normal(0, 3, n_sellers)  # smaller individual noise
)

# -----------------------------
# Fit mixed-effects model
# -----------------------------
model_mixed = smf.mixedlm('Revenue ~ Strategy + Experience', exp_df, groups=exp_df['Region'])
result = model_mixed.fit(reml=False, method='lbfgs', maxiter=1000)
print(result.summary())
```

In this experimental study, sellers were randomly assigned to marketing strategies (A, B, or C) across 20 regions, and revenue was analyzed using a mixed-effects model with Region as a random intercept. The model accounts for both fixed effects (Strategy and Experience) and random variation across regions, capturing baseline differences in revenue between regions. The results indicate that Strategy B increased revenue by an average of 4.82 units and Strategy C by 2.44 units, both statistically significant (p < 0.001) compared to Strategy A. Experience also had a strong positive effect, with revenue increasing by about 2 units per year of experience. The estimated variance of the random intercept for Region was 66.64, indicating meaningful variability in baseline revenue across regions. Overall, this mixed-effects approach appropriately models the nested structure of the data and provides robust estimates of strategy effects while accounting for regional differences

#### **Causal Mediation Analysis**

Causal Mediation Analysis helps us understand not just whether a marketing strategy affects revenue, but how it does so. In many cases, the effect of a strategy may operate through intermediate variables—called mediators—rather than solely directly. In this study, we considered the seller's MarketingBudget as a potential mediator. Sellers with more experience tend to allocate higher budgets, which could amplify the impact of the chosen strategy on revenue.

We first modeled how the strategy influences the marketing budget and then how both strategy and marketing budget affect revenue. By multiplying these effects, we estimated the indirect effect—the portion of the strategy's impact on revenue that occurs through changes in marketing budget. This allows us to distinguish the direct effect of the strategy (impact not explained by budget) from the indirect effect (impact transmitted via budget). For example, Strategy B increases revenue partly because it leads sellers to allocate more budget, illustrating the mechanism behind the observed revenue gains.

Overall, causal mediation analysis provides a more nuanced understanding of marketing strategy effectiveness, revealing both direct and mediated pathways that contribute to revenue outcomes.

```python
exp_df['MarketingBudget'] = exp_df['Experience']*10 + np.random.normal(0,50,n_sellers)

med_model = smf.ols('MarketingBudget ~ Strategy', data=exp_df).fit()
out_model = smf.ols('Revenue ~ Strategy + MarketingBudget', data=exp_df).fit()

indirect_effect = med_model.params['Strategy[T.B]'] * out_model.params['MarketingBudget']
indirect_effect
```

The causal mediation analysis shows that the indirect effect of Strategy B on revenue through marketing budget is small and slightly negative (-0.73), indicating that the mediated pathway does not contribute positively to revenue. This suggests that any effect of Strategy B on revenue occurs primarily through direct rather than budget-mediated channels.

#### **Randomization Inference**

Randomization inference is a non-parametric method that leverages the actual random assignment of treatments to evaluate the effect of an intervention. Instead of relying on assumptions about the distribution of errors (as in traditional t-tests), it directly tests whether the observed difference in outcomes between treatment groups could plausibly occur by chance under random assignment.

In this example, we compared revenue between sellers assigned to Strategy A and Strategy B. The observed difference in mean revenue was calculated, and then a large number of random permutations were generated by shuffling the group labels while keeping group sizes fixed. For each permutation, the difference in means was recalculated, forming a distribution of differences expected under the null hypothesis of no treatment effect. The permutation p-value is the proportion of these permuted differences that are as extreme as or more extreme than the observed difference. A small p-value indicates that the observed difference is unlikely to arise by chance, providing strong evidence that the treatment (strategy) has a causal effect on revenue.

This approach is particularly appealing in experimental studies because it directly reflects the randomization design and does not rely on parametric assumptions, making it robust and conceptually straightforward for testing causal effects

```python
from itertools import permutations

group_A = exp_df[exp_df['Strategy']=='A']['Revenue'].values
group_B = exp_df[exp_df['Strategy']=='B']['Revenue'].values
obs_diff = np.mean(group_B) - np.mean(group_A)

perm_diffs = []
for _ in range(1000):
  combined = np.concatenate([group_A, group_B])
  np.random.shuffle(combined)
  perm_diffs.append(np.mean(combined[:len(group_B)]) - np.mean(combined[len(group_B):]))
perm_diffs = np.array(perm_diffs)
p_value_perm = np.mean(np.abs(perm_diffs) >= np.abs(obs_diff))
print("Permutation p-value:", p_value_perm)
```

The randomization inference shows a permutation p-value of 0.0, indicating that the observed increase in revenue for Strategy B compared to Strategy A is extremely unlikely to occur by chance. This provides strong evidence that Strategy B has a significant causal effect on revenue

## **Quasi-Experimental Design**

Quasi-experiments are study designs where the researcher cannot fully randomize treatment assignment, but exploits some natural or policy-driven variation to estimate causal effects.

Key Features:

- No full randomization, but some as-if random assignment may exist.
- Often relies on external shocks, policy changes, or thresholds.
- Requires careful design and statistical methods to control for confounding.

Common Methods in Quasi-Experiments:

- Difference-in-Differences (DiD): Compares changes in outcomes over time between treated and control groups.
- Regression Discontinuity (RD): Exploits a cutoff or threshold that determines treatment assignment.
- Instrumental Variables (IV): Uses a variable that affects treatment but not directly the outcome (already discussed in observational studies, but often applied here).
- Synthetic Control Methods: Creates a weighted combination of control units to match treated units pre-intervention.
- Interrupted Time Series (ITS): Analyzes before-and-after trends around a specific intervention point to estimate both immediate level changes and long-term trend shifts.

Use Cases:

- Evaluating the effect of a new policy rolled out in some regions but not others.
- Studying marketing campaigns launched at specific sales thresholds.
- Measuring the impact of regulation changes or natural experiments.

### **Difference-in-Differences (DiD)**

Difference-in-Differences is one of the most widely used tools in quasi-experimental research because it lets us estimate causal effects even when we cannot run a true experiment. The idea is simple: instead of only comparing treated and untreated groups after a policy or intervention, DiD compares how each group changes over time. If the treated group improves more than the control group over the same period, that extra improvement can be attributed to the treatment—assuming that without the intervention, the two groups would have followed similar trends.

In marketing terms, imagine launching a new promotion in one region (East) but not another (West). Both regions already have different baseline revenue levels, and both may naturally grow over time. DiD handles these issues by focusing on differences in growth, not raw levels. If the East sees a bigger jump after the promotion than the West does, that additional boost is interpreted as the causal impact of the promotion.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

np.random.seed(42)
n_sellers = 100

# Generate data
df = pd.DataFrame({
    'Seller': range(n_sellers),
    'Region': np.random.choice(['East', 'West'], size=n_sellers),
    'PreRevenue': np.random.normal(100, 10, n_sellers)
})

# Treatment: East region gets new promotion
df['PostRevenue'] = df['PreRevenue'] + np.where(df['Region']=='East', 15, 2) + np.random.normal(0,5,n_sellers)

# Create dummy variables
df['Treatment'] = (df['Region']=='East').astype(int)
df['Post'] = 1  # after intervention
df['Interaction'] = df['Treatment'] * df['Post']

# DiD regression
model = smf.ols('PostRevenue ~ Treatment + Post + Interaction', data=df).fit()
print(model.summary())
```

The Difference-in-Differences regression shows a strong and statistically significant treatment effect. The key coefficient—the Interaction term—is positive (7.35) and highly significant (p < 0.001), indicating that the East region’s revenue increased by about 7.3 units more than the West region after the promotion. This suggests that the promotion produced a meaningful causal lift in revenue beyond normal trends or baseline differences.

### **Regression Discontinuity (RD)**

Regression Discontinuity is a quasi-experimental method that takes advantage of a rule or cutoff that determines who gets a treatment. The key idea is simple: units just above and just below a threshold are nearly identical in all respects except that one group receives the treatment. Because of this, any sudden jump in outcomes at the cutoff can be interpreted as a causal effect of the treatment.

Marketing Use Case:
Imagine a company gives access to a premium marketing tool only to sellers with 5 or more years of experience. Sellers with 4.9 years are almost indistinguishable from sellers with 5.1 years in terms of skills and business maturity, but only the latter get the tool. RD compares revenue right around that cutoff. If we see a sharp increase in revenue exactly at 5 years of experience, it provides strong evidence that the marketing tool itself and not unrelated seller characteristics caused the improvement.

```python
np.random.seed(42)
n_sellers = 100

df_rd = pd.DataFrame({
    'Experience': np.random.uniform(1,10, n_sellers),
})
df_rd['Treatment'] = (df_rd['Experience'] >= 5).astype(int)
df_rd['Revenue'] = 50 + 5*df_rd['Experience'] + 10*df_rd['Treatment'] + np.random.normal(0,5,n_sellers)

# RD regression with linear term and treatment
model_rd = smf.ols('Revenue ~ Experience + Treatment', data=df_rd).fit()
print(model_rd.summary())
```

The regression discontinuity analysis shows a clear jump in revenue at the 5-year experience cutoff. Sellers who qualify for the premium marketing tool earn about 8.5 units more revenue, even after accounting for the continuous effect of experience. Both the treatment effect and the experience slope are highly significant, indicating that the premium tool provides a meaningful boost in performance beyond what experience alone would generate

### **Interrupted Time Series (ITS)**

Interrupted Time Series (ITS) is a powerful quasi-experimental method used when we observe outcomes repeatedly over time and a clear intervention or event occurs at a specific point. Instead of simply comparing before and after averages, ITS explicitly models the underlying trend leading up to the intervention and asks whether the event caused (1) an immediate jump in the outcome or (2) a change in the trajectory afterward. This makes ITS especially useful in marketing and policy settings where natural time trends like seasonal patterns or gradual growth must be accounted for.

Marketing Example:
Suppose a company launches a major campaign in January 2025. Revenue naturally grows a little each month, so simply seeing higher revenue after January doesn't prove the campaign worked. ITS solves this by modeling the pre-campaign trend and then checking whether the campaign caused a statistically detectable change in the level of revenue or its growth rate. If we see a sudden upward shift right after the intervention—beyond what the trend predicted that is strong evidence of a campaign effect.

```python
months = pd.date_range('2023-01-01', periods=24, freq='M')
n = len(months)

np.random.seed(42)
revenue = 100 + np.arange(n)*2 + np.random.normal(0,5,n)  # baseline trend
intervention = np.array([0]*12 + [1]*12)
revenue_post = revenue + intervention*10  # effect of campaign

df_its = pd.DataFrame({
    'Month': months,
    'Revenue': revenue_post,
    'Time': np.arange(n),
    'Intervention': intervention
})

# ITS regression
model_its = smf.ols('Revenue ~ Time + Intervention', data=df_its).fit()
print(model_its.summary())
```

The Interrupted Time Series regression shows a strong positive underlying revenue trend over time, as reflected in the significant coefficient on the time variable. Importantly, the intervention term is also positive and statistically significant, indicating that the marketing campaign produced an immediate upward shift in revenue beyond what the existing trend would have predicted. This suggests that the campaign had a meaningful and measurable impact on monthly revenue levels

### **Synthetic Control Method**

The Synthetic Control Method is a powerful tool used when we want to estimate the causal effect of an intervention applied to a single treated unit, such as one region, one store, or one country. Instead of relying on a traditional control group, SCM builds a “synthetic” version of the treated unit by combining multiple control units using optimized weights. The idea is that this synthetic control closely matches the treated unit's pre-intervention behavior, giving us a more credible counterfactual for what would have happened without the treatment.

Marketing Example: Suppose a company runs a major promotion only in the East region. To measure its impact, we need to know what East's revenue would have looked like had the promotion not occurred. Instead of choosing a single comparison region, SCM blends other regions—such as West and Central—in the proportions that best mimic East before the promotion. This synthetic control then serves as the baseline for comparison.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


np.random.seed(42)
months = pd.date_range('2023-01-01', periods=12, freq='ME')

# Create revenue for 3 regions: East (treated), West, Central
data = pd.DataFrame({
    'Month': np.tile(months, 3),
    'Region': ['East']*12 + ['West']*12 + ['Central']*12,
    'Revenue': np.nan
})

# Generate baseline revenue trends
baseline = 100 + np.arange(12)*2
data.loc[data['Region']=='East', 'Revenue'] = baseline + np.random.normal(0, 2, 12)
data.loc[data['Region']=='West', 'Revenue'] = baseline*0.95 + np.random.normal(0, 2, 12)
data.loc[data['Region']=='Central', 'Revenue'] = baseline*1.05 + np.random.normal(0, 2, 12)

# Introduce treatment effect: East gets promotion in month 7-12
data.loc[(data['Region']=='East') & (data['Month']>=months[6]), 'Revenue'] += 10

# 2. Prepare matrices for SCM

pre_period = months[:6]  # months before intervention
treated_pre = data[(data['Region']=='East') & (data['Month'].isin(pre_period))]['Revenue'].values

controls = ['West', 'Central']
control_pre_matrix = np.column_stack([
    data[(data['Region']==ctrl) & (data['Month'].isin(pre_period))]['Revenue'].values
    for ctrl in controls
])

# 3. Find optimal weights (non-negative, sum=1)

from scipy.optimize import minimize

def loss(weights):
  synthetic = control_pre_matrix @ weights
  return np.sum((treated_pre - synthetic)**2)

# initial guess
init_weights = np.array([0.5, 0.5])
bounds = [(0,1)]*len(controls)
constraints = {'type':'eq', 'fun': lambda w: np.sum(w)-1}

res = minimize(loss, init_weights, bounds=bounds, constraints=constraints)
weights_opt = res.x
print("Optimal weights for synthetic control:", dict(zip(controls, weights_opt)))


# 4. Construct synthetic control for all months

control_matrix_full = np.column_stack([
    data[data['Region']==ctrl]['Revenue'].values
    for ctrl in controls
])
synthetic_control = control_matrix_full @ weights_opt


# 5. Estimate treatment effect

treated_full = data[data['Region']=='East']['Revenue'].values
effect = treated_full - synthetic_control

data_scm = pd.DataFrame({
    'Month': months,
    'Treated': treated_full,
    'Synthetic': synthetic_control,
    'Effect': effect
})

print("Average estimated causal effect (post-intervention):",
      np.mean(effect[6:]))  # post-intervention months
```

The synthetic control analysis shows that the optimal weighted combination of the control regions assigns approximately 34% to West and 66% to Central, creating a counterfactual for the treated East region. Comparing actual East revenue to this synthetic control after the promotion, the estimated causal effect is about 8.91, indicating that the marketing intervention increased revenue by roughly 9 units on average
